4) What happens if you map the same variable to multiple aesthetics?
There both displayed 

```r
ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy, color = cyl, size = cyl))
```

```r
ggplot(data=mpg) + geom_point(mapping=aes(x=displ, y=hwy, color= drv, shape=drv))
```

5) What does the stroke aesthetic do? What shapes does it work with?
It increases the sizes of the shapes

```r
ggplot(data=mpg) +
  geom_point(mapping = aes(x=displ, y=hwy, color=cyl, size=hwy, shape= drv, stroke=5))
```

6) What happens if you map an aesthetic to something other than a variable name, like aes(colour = displ < 5)? 
It applies to the evaluated value 

```r
ggplot(data = mpg) + geom_point(mapping = aes(x = displ, y = hwy, colour = displ < 5))
```
